```python
import random
import time
```

```python
# Producer
def producer(n=10):
    for i in range(n):
        yield {"event_id": i, "value": random.randint(1, 100)}
```

```python
# Consumer processing pipeline
def process_event(event):
    event["processed_value"] = event["value"] * 2
    return event

stream = producer(20)
```

```python
processed_stream = [process_event(e) for e in stream]

processed_stream[:5]
```
